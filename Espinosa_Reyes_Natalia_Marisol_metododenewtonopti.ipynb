Codigo para metodo de newton.

```python
import numpy as np

def grad(x):
    x1, x2 = x
    return np.array([
        -400*x1*(x2 - x1**2) + 2*(x1 - 1),
        200*(x2 - x1**2)
    ])

def hessiana(x):
    x1, x2 = x
    return np.array([
        [1200*x1**2 - 400*x2 + 2, -400*x1],
        [-400*x1, 200]
    ])

x = np.array([-1.2, 1.0])

for k in range(2):  # dos iteraciones
    p = np.linalg.solve(hessiana(x), -grad(x))
    x = x + p
    print(f"Iteración {k+1}: x^* = {x}")

print("\nResultado final:", x)
```
